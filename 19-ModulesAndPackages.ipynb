```python
import numpy as np
import matplotlib.pyplot as matplot
```

```python
maasListesi = np.random.normal(4000,500,1000)
numpy.mean(maasListesi)     # numpy: verilerin oluşturulmasını sağlar
```

```python
matplot.hist(maasListesi,50)
matplot.show()                       # matplot: verilerin görselleştirilmesini sağlar
```
